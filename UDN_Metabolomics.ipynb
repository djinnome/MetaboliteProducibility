## Add names to genes in Recon3D

```python

def add_name_to_genes( model, gene_names ):
    for gene in model.genes:
        if gene.id in gene_names:
            gene.name = gene_names[gene.id]
    return model


recon3d = cobra.io.load_json_model('Recon3D_301/Recon3DModel_301.json')
gene_names = pd.read_csv('ncbi-gene-to-common-name.tab',sep='\t',index_col='NCBI').to_dict()['CommonName']
recon3d = add_name_to_genes(recon3d, gene_names)

cobra.io.save_json_model(recon3d, 'Recon3D_301/Recon3DModel_301.names.json', pretty=True)
```

```python
from cobra.flux_analysis import single_gene_deletion
import cobra

def mets_of_bigg_names( model, bigg ):
    metabolites = []
    for bigg_id in bigg.index:
        for met in model.metabolites.query(bigg.loc[bigg_id,
                                                      'Description:bigg'], 
                                         'name'):
            metabolites.append(met)
    return metabolites


def gene_name( model, gene_id ):
    if model.genes.has_id(gene_id):
        gene = model.genes.get_by_id(gene_id)
        if gene.name != '':
            return gene.name
        else:
            return gene.id
    else:
        return gene_id
    
def get_kos_of_metabolite_producibility( metabolite_producibility, threshold = 0.98):
    kos = {}
    for met in metabolite_producibility.columns:
        ko_growth = metabolite_producibility[metabolite_producibility[met] < threshold][met].to_dict()
        if ko_growth != {}:
            kos_of_gene = dict( [(gene_name(recon3d, k), v  ) 
                                   for k,v in ko_growth.items()] )
            kos['{}:{}'.format(met, 
                               recon3d.metabolites.get_by_id(met).name)] = kos_of_gene
    return kos

def run_metabolite_producibility( model, metabolites,threshold=0.98  ):
    metabolite_producibility = pd.DataFrame()
    for met in metabolites:
        with model:
            demand_rxn = 'DM_{}'.format(met.id)
            if not model.reactions.has_id( demand_rxn ):
                model.add_boundary(met, type='demand')
            model.objective = demand_rxn
            wt_growth = model.slim_optimize()
            metabolite_producibility[met.id] = single_gene_deletion( model ).growth/wt_growth
    metabolite_producibility.index = [list(gene)[0] 
                                          for gene in metabolite_producibility.index]  
    
    return get_kos_of_metabolite_producibility( metabolite_producibility, threshold )

recon3d = cobra.io.load_json_model('Recon3D_301/Recon3DModel_301.names.json')
metabolites = mets_of_bigg_names( recon3d, udn_bigg )
metabolite_kos = run_metabolite_producibility( recon3d, metabolites )       
with open('metabolite_kos.json', 'w') as out:
    json.dump( metabolite_kos, out)
metabolite_kos
```
