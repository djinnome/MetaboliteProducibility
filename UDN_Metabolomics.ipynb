# Map UDN Metabolites to MetaCyc

```python
pnnl_to_metacyc = pd.read_table('UDN_Metabolites_PNNL_to_Metacyc.tab',index_col='PNNL Metabolite Name (Urine & Plasma) ')
pnnl_to_metacyc.loc['nonoic acid'] = pd.Series({'Compound':'CPD-8505', 'Metacyc Object ID':'CPD-8505'} )
pnnl_to_metacyc.loc['carbote ion'] = pd.Series({'Compound':'CO3', 'Metacyc Object ID':'CO3'})
pnnl_to_metacyc.loc['1-monostearin*'] = pd.Series({'Compound':'CPD-16717', 'Metacyc Object ID':'CPD-16717'})
pnnl_to_metacyc.loc['malonic acid 2 (deuterated?)'] = pd.Series({'Compound':'MALONATE', 'Metacyc Object ID':'MALONATE'})
pnnl_to_metacyc.loc['2-hydroxybutyric acid'] = pd.Series({'Compound':'CPD-3564', 'Metacyc Object ID':'CPD-3564'})
```

```python
import numpy as np
udn = pd.read_table('Unique_UDN_Plasma_Pilot_thr_Project06_zscore_METAB_v2.txt').join(pnnl_to_metacyc, on='Metabolite',how='inner').set_index('Compound')
for i in range(4):
    udn.iloc[:,i*50:(i+1)*50].to_csv('Unique_UDN_Plasma_Pilot_thr_Project06_zscore_METAB_v2_{}.tab'.format(i),sep='\t')
udn
```

```python
print('\n'.join(sorted(udn[udn['Metacyc Object ID'].isnull()].index)))
```

```python
with open('udn-cpds.lisp','w') as out:
    out.write('''(setq *udn-cpds* '({}))'''.format('\n'.join(sorted(udn[udn['Metacyc Object ID'].notnull()]['Metacyc Object ID'].values))))
```

# Map MetaCyc IDs to Bigg using MetaNetX 

```python
chemxref = 'https://www.metanetx.org/cgi-bin/mnxget/mnxref/chem_xref.tsv'
import metanetx
cyc2bigg = metanetx.convert_db('metacyc','bigg').set_index('metacyc')
cyc2bigg['bigg']
```

```python
cyc2bigg.drop_duplicates('Description:bigg')
```

```python
udn_bigg = udn[udn.columns[-9:]].join(cyc2bigg.drop_duplicates('Description:bigg'), on='Metacyc Object ID', how='inner').reset_index().set_index('bigg')
udn_bigg.to_csv('UDN_cpds_bigg.tab',
                sep='\t')
print('\n'.join(udn_bigg.columns))
```
